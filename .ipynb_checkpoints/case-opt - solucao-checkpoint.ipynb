#### Otimização Plano de Produção:
##### Restrições:
###### Demanda
Um empresa, fabricante de bicicletas deseja planejar a produção de 10 dos seus principais produtos (SKU), $SKU_{001}, SKU_{002}, ..., SKU_{010}$ em um horizonte de 4 meses, de janeiro a dezembro.

Dentro da demanda mensal existem pedidos feitos por uma grande rede varejista que deve ser preferencialmente atendida, sob risco de prejudicar o relacionamento com o cliente. Esse varejista exige que o pedido seja atendido em sua totalidade, caso contrário cancela o pedido total. Ou seja, se o total de bicicletas solicitado foi 30, ou entrega-se as 30 ou o pedido inteiro é cancelado.

**Base de Demanda e Pedidos de Varejistas**
- Arquivo: `demand.csv`
- Colunas: `sku`: Código do Produto | `month`: Mês | `demand`: Demanda no mês (# bicicletas) | `retail`: Pedido firme da rede varejista

###### Disponibilidade de Matéria Prima
Por se tratar de uma pequena fábrica, a disponibilidade de matéria prima não é uma restrição, embora seu custo esteja embutido no custo do produto.

**Base custo variável e Preço de venda**
- Arquivo: `costs_revenues.csv`
- Colunas: `sku`: Código do Produto | `month`: Mês | `cost_unit`: Custo por unidade produzida (Mão de obra mais MP) | `price`: Preço de venda do produto acabado

###### Linha de produção
Algumas linhas de produção podem produzir mais de um tipo de bicicleta, enquado outras são monoproduto. Cada linha de produção incorre em um custo fixo cada vez que é acionada (devido a manutenção e custo de partida dos motores), independente da quantidade produzida.

**Possibilidades de produção e custo fixo**
- Arquivo: `production_arcs.csv`
- Colunas: `machine`: Linha de produção | `sku`: Código do Produto | `eficiency`: Tempo de produção de uma unidade unid/min
------------------
- Arquivo: `activation_cost.csv`
- Colunas:  `machine`: Linha de produção | `activation_cost`: Custo de Ativação da linha |

**Capacidade produtiva**
- Arquivo: `production_capacity.csv`
- Colunas: `machine`: Linha de produção | `capacity`: Capacidade mensal da máquina (contando horas de manutenção e etc), em h / mês

###### Subprodutos
O processo produtivo gera alguns subprodutos que podem ser vendidos (por Kg) para empresas de reciclagem, doados para cooperativas ou descartados, sendo que o custo de transporte e a capacidade de processamento são diferentes em cada caso (Não leva-se em consideração o impacto social das decisões).

**Geração de subprodutos**
- Arquivo: `by_products.csv`
- Colunas: `machine`: Linha de produção | `sku`: Código do Produto | `by_product`: Kg gerados por unidade produzida

**Capacidade de escoamento de subprodutos**
- Arquivo: `by_products_capacity.csv`
- Colunas: `customer`: Cooperativa, reciclagem ou lixo  | `capacity`: Demanda | `price`: Preço pago (somente a empresa de reciclagem paga algo pelo material) R\$/KG | `cost`: Caso o produto tenha que ser jogado fora, existe um custo de transporte R\$/Kg.

###### Capacidade de estoque
A fabrica possui um estoque próprio, que pode não ser suficiente absorver todo sua produção. No entanto é possível alugar espaço em um galpão ao lado da fábrica. Existe um acordo entre o dono do galpão e a fábrica de bicicletas onde é pago um custo fixo por unidade estocada (Como as bicicletas têm tamanho semelhante, a questão de volume se torna irrelevante). Como os prédios estão muito próximos, o custo de transporte é irrelevante.

**Capacidade de estoque**
- Arquivo: `storage_capacity.csv`
- Colunas: `unit`: Unidade de armazenagem  | `capacity`: Capacidade em unidade de bicicletas | `price`: Custo de estocagem por unidade R$ / Unid

# 1 - Classificação do problema

É um problema de otimização que pode ser modelado como um problema de programação linear inteira mista.

Ou seja, as variáveis po modelo tem variáveis que assumem valores inteiros, reais (contínuos) e binários.

Todas as restrições são lineares.

Há duas funções objetivo conflitantes, maximizar o lucro e minimizando o atrito.

# 2 - Como evitar atritos nas soluções finais?

Para tentar minimizar os atritos com o cliente podemos ver como um problema multiobjetivo e avaliar os trade-offs com a quantidade de pedidos não atendidos. Podemos trazer isso de diversar maneiras. Neste case tratarei a função objetivo como a soma de duas funções: 1. o lucro total, 2. um custo fixo para pedidos não atendidos. Cada função tem um parâmetro de controle, e soluções ótimas de pareto podem ser analizadas.

O uso de programação multiobjetivo dos próprios solvers também pode ser avaliadas.

Também podemos limitar o número máximo de pedidos não atendidos com uma tolerância "aceitável". E adicionar isso como uma restrição do modelo.

Ou também, podemos usar programação multinível fixando um número de pedidos aceitos dados pelo máximo possível de pedidos aceitos do modelo e apartir disso resolver outro modelo maximizando o lucro.


# 3 - Modelo matemático de programação linear inteira mista para o problema


## Dados do problema

$N$ : conjunto de produtos

$M$ : conjunto de máquinas (linhas de produção)

$P$ : conjunto de pedidos. Considerando que cada pedido tem um par (i,t) único, ou seja, não considerei que um pedido por ter demanda por mais de um item no mesmo período. Para considerar a variável de decisão q e matriz de demandas aumentariam uma dimensão.

$T$ : conjunto de períodos (meses)

$U$ : conjunto de unidades de armazenagem

$V$ : conjunto de clientes destino para os subprodutos (empresas de reciclagem, doados para cooperativas ou descartados)

${\cal I}_k$ : conjunto de produtos que podem ser produzidos por $k \in M$

${\cal F}_i$ : conjunto de máquinas que podem produzir $i \in N$

$d_{ijt}$ : demanda do produto $i \in N$ no mês $t \in T$ associada ao pedido $j \in P$ (unidade). 

$c^{mo\&mp}_{it}$ : custo de mão de obra mais matéria prima do produto $i \in N$ no mês $t \in T$ (R$ por unidade)

$p_{it}$ : preço de venda do produto $i \in N$ no mês $t \in T$ (R$ por unidade)

$c^{ativ}_k$: Custo de ativação da máquina $k \in M$ (R$)

$c^{alu}_u$: Custo de aluguel da unidade da armazenamento $u \in U$ (R$ por unidade)

$p^{sub}_v$ : Preço de venda de subprodutos para o cliente $v \in V$. (R$ por quilo) Caso deseje-se implementar tipos de subprodutos diferentes esses parâmetros e variáveis associado aos sub produtos aumentariam uma dimensão. 

$c^{sub}_v$ : Custo de transporte ou outros custos eventuais de vender/doar subprodutos. (R$ por quilo)

$sub_{ik}$: subproduto gerado por unidades produzida de $i \in N$ por $k \in {\cal F}_{i}$. (quilos)

$e_{ik}$ : Eficiência da máquina $k \in {\cal F}_i$ para produzir o item $i \in N$ (minutos)


$cap^{prod}_{kt}$ : capacidade de produção da máquina $k \in M$ no período $t \in T$ em horas (para a instância é fixo por máquina, então basta $cap^{prod}_{k}$ para simplificar) (unidades)

$cap^{sub}_{v}$ : capacidade/demanda de subprodutos por cliente $v \in V$. (quilos)

$cap^{armaz}_{u}$ : capacidade da unidade de armazenagem $u \in U $ em unidades, inclui o próprio armazem (unidades)

${\cal M}$ : um número suficientemente grande, pode ser estimado com base na demanda


## Variáveis de decisão

### Note que algumas variáveis podem ser relaxadas como reais, já que são obtidas por operações lineares entre inteiros e consequentemente serão inteiras na solução

$q_{ikt} \in \mathbb{R}^+$ : quantidade de produtos do tipo $i \in N$ produzidos pela máquina $k \in M$ no período $t \in T$ em unidades. 

$y_{ijt} \in \mathbb{B}^+$ : 1, se a demanda do pedido  $j \in P$  de produto $i \in N$ do período $t$ é atendido, 0, caso contrário

$w_{kt} \in \mathbb{B}^+$ : 1, se a máquina $k \in M$ é usada no período $t \in T$, 0, caso contrário

$I_{it} \in \mathbb{R}^+$ : estoque de item $i \in N$ no final do período $t \in T$ em unidades, note que para esse problema estoque inicial $I_{i0} = 0$.

$I_{ut} \in \mathbb{R}^+$ : estoque total utilizado na unidade de armazenagem $u \in U$ no final do período $t \in T$ em unidades, note que para esse problema estoque inicial $I_{u0} = 0$.

$z_{u t} \in \mathbb{B}^+$ : 1 se a unidade de armazenagem $u \in U$ é alugada no período $t \in T$, 0 caso contrário

$q^{sub}_{vt} \in \mathbb{R}^+$ : peso em kg de subprodutos gerados para o cliente $v \in V$ no final do período $t \in T$

## Restrições

### Restrição de conservação de demanda:
#### Estoque dos itens no final do mês anterior + quantidade produzida - o estoque no final do mês tem que ser igual a demanda dos pedidos atendidos:

$\sum_{u \in U} I_{iu,t-1} + \sum_{k \in {\cal F}_i} q_{ikt} - \sum_{u \in U} I_{iut} = \sum_{j \in P} y_{ijt} d_{ijt}, \quad i \in N, t \in T, j \in P$

### Restrição de ativação das máquinas:
#### Uma máquina é ativa no período se e somente se há produção de algum item nela neste mesmo período:

$\sum_{i \in {\cal I}_{k}} q_{ikt} \leq {\cal M} \cdot w_{kt}, \quad \forall k \in M, t \in T$

### Restrição de capacidade de produção
#### para cada máquina e período o tempo total em minutos da produção dos itens é melhor ou igual a capacidade total da máquina em minutos:

$ \sum_{i \in {\cal I}_k} e_{ik} q_{ikt} \leq 60\cdot cap^{prod}_k, \quad k \in M, t \in T$

### Restrição de estoque
### para cada período o estoque no final do período não é maior que a sua capacidade de armazenamento, somando toda a capacidade alugada das unidades de armazenamento

$I_{ut} = \sum_{i \in N} I_{iut}, \quad t \in T, u \in U $

$I_{ut} \leq cap^{armaz}_{u} \cdot z_{ut}, \quad t \in T, u \in U $

### Restrições de capacidade de subprodutos
### Definição da quantidade total de quilos de subprodutos destinados a cada cliente

$ \sum_{v \in V} q^{sub}_{vt} = \sum_{i} \sum_{k} sub_{ik} \cdot q_{ikt}, t \in T$

### A quantidade total de quilos de subprodutos destinados a cada cliente não deve exceder sua capacidade/demanda por período

$ q^{sub}_{vt} \leq cap^{sub}_v, \quad v \in V, t \in T$

Naturalmente o descarte só é feito se a capacidade de reciclagem ou doação é excedida, já que existe um custo de transporte, nos outros casos há ganho por venda ou custo zero. Como o modelo é de maximização de lucro isso só ocorrerá se as capacidades forem excedidas no período

## Função objetivo

## Inicialmente maximizar o lucro total

#### Lucros de venda de produtos + lucro de vendas de subprodutos - custos de armazenagem - custos de ativação de máquinas

maximizar
$f_1(x) = \sum_{i \in N} \sum_{k \in {\cal F}_i} \sum_{t \in T} (p_{it} - c^{mo\&mp}_{it}) y_{ijt} d_{ijt}+ \sum_{v \in V}\sum_{t \in T} (p^{sub}_{v} - c^{sub}_v) \cdot q^{sub}_{vt}- \sum_{u \in U} \sum_{t \in T} c^{alu}_u I_{ut}- \sum_{k \in M} \sum_{t \in T} c^{ativ}_k w_{kt} $

#### Para tentar minimizar os atritos com o cliente podemos ver como um problema multiobjetivo e avaliar os trade-offs com a quantidade de pedidos não atendidos.

Podemos trazer isso de diversar maneiras. Neste case tratarei a função objetivo como

maximizar $ f(.) = \alpha f_1(.) - \beta f_2(.) $

Onde $f_2(.) = \sum_{i \in N}\sum_{j \in P}\sum_{t \in T} (1 -y_{ijt})$ descreve o número total de pedidos não atendidos. $\alpha$ e $\beta$ são parâmetros de controle onde podemos ponderar a preferência entre lucro e números de pedido aceitos.

Para somar unidades iguais podemos definir um custo em reais estimado para cada pedido não aceito e podemos reescrever $f_2(x)$ como 

$f_2(x) = \sum_{i \in N}\sum_{j \in P}\sum_{t \in T} c^{atrito}_{ijt} (1 -y_{ijt})$ com $c^{atrito}_{ijt}$ representando um custo de atrito por não atender o pedido $j$


# 4 - Implementação do modelo com Pyomo e CPLEX

#### Carregar os dados:

```python
#importar a biblioteca pandas
import pandas as pd
#ler dados e colocar eles em um pandas dataframe
demand = pd.read_csv('./data/demand.csv'.format('csv'), sep=';', decimal=',', encoding='utf-8')
production_arcs = pd.read_csv('./data/production_arcs.csv'.format('csv'), sep=';', decimal=',', encoding='utf-8')
production_capacity = pd.read_csv('./data/production_capacity.csv'.format('csv'), sep=';', decimal=',', encoding='utf-8')
cost_revenues = pd.read_csv('./data/cost_revenues.csv'.format('csv'), sep=';', decimal=',', encoding='utf-8')
by_products_capacity = pd.read_csv('./data/by_products_capacity.csv'.format('csv'), sep=';', decimal=',', encoding='utf-8')
by_products = pd.read_csv('./data/by_products.csv'.format('csv'), sep=';', decimal=',', encoding='utf-8')
activation_cost = pd.read_csv('./data/activation_cost.csv'.format('csv'), sep=';', decimal=',', encoding='utf-8')
storage_capacity = pd.read_csv('./data/storage_capacity.csv'.format('csv'), sep=';', decimal=',', encoding='utf-8')
```

#### Montar o modelo em si

```python
import pyomo.environ as pyo
import numpy as np

def MontarModelo(demand,
                 production_arcs,
                 production_capacity,
                 cost_revenues,
                 by_products_capacity,
                 by_products,activation_cost,
                 storage_capacity,
                 isMultiObjective = True,
                 hasLimitedProduction = True,
                 hasLimitedStock = True,
                 alpha = 1,
                 beta = 0
                ):
    
    #Conjuntos
    PRODUTOS = demand['sku'].unique()
    PERIODOS = demand['month'].unique()
    PEDIDOS = demand['retail'].unique()
    MAQUINAS = production_arcs['machine'].unique()
    DESTINOS = by_products_capacity['customer'].unique()
    UNIDADESDEARMAZEM = storage_capacity['unit'].unique()
    
    #Dados
    demand = demand.set_index(['retail','sku','month'])
    existe = np.zeros(shape=(max(PEDIDOS),len(PRODUTOS), len(PERIODOS)))
    for idx in demand.index:
        existe[idx[0]-1][idx[1]-1][idx[2]-1] = 1

    production_arcs = production_arcs.set_index(['sku', 'machine'])
    activation_cost = activation_cost.set_index('machine')
    cost_revenues= cost_revenues.set_index(['sku','month'])
    by_products_capacity = by_products_capacity.set_index(['customer',])
    by_products = by_products.set_index(['sku', 'machine'])
    storage_capacity = storage_capacity.set_index('unit')
    production_capacity = production_capacity.set_index('machine')
    BIG_M = 99999999
    
    # Objeto do modelo
    model = pyo.ConcreteModel()
    
    #Definição de variáveis de decisão
    model.q = pyo.Var(production_arcs.index, PERIODOS, domain = pyo.NonNegativeReals)
    model.y = pyo.Var(demand.index, domain = pyo.Binary)
    model.w = pyo.Var(MAQUINAS, PERIODOS, domain = pyo.Binary)
    model.I = pyo.Var(PRODUTOS, UNIDADESDEARMAZEM, PERIODOS, domain = pyo.NonNegativeReals)
    model.I_L = pyo.Var(UNIDADESDEARMAZEM, PERIODOS, domain = pyo.NonNegativeReals)
    model.z = pyo.Var(UNIDADESDEARMAZEM, PERIODOS, domain = pyo.Binary)    
    model.q_sub = pyo.Var(DESTINOS, PERIODOS, domain = pyo.NonNegativeReals)
    
    model.custo_ativacao = pyo.Var(domain = pyo.NonNegativeReals)
    model.custo_estoque = pyo.Var(domain = pyo.NonNegativeReals)
    model.lucro_venda = pyo.Var(domain = pyo.NonNegativeReals)
    model.lucro_subprodutos = pyo.Var(domain = pyo.NonNegativeReals) 
    model.custo_atrito = pyo.Var(domain = pyo.NonNegativeReals)
    
    #Montar função objetivo
    custo_ativacao = sum(activation_cost.loc[k]['activation_cost'] * model.w[k,t] for k in MAQUINAS for t in PERIODOS)
    custo_estoque = sum(model.I_L[l, t] * storage_capacity.loc[l]['price'] for l in UNIDADESDEARMAZEM for t in PERIODOS)
    
    lucro_venda = 0
    for i in PRODUTOS:
        for j in PEDIDOS:
            for t in PERIODOS:
                if existe[j-1][i-1][t-1] == 1:
                    lucro_venda += (cost_revenues.loc[i,t]['price'] - cost_revenues.loc[i,t]['cost_unit']) *model.y[j,i,t] * demand.loc[j,i,t]['demand'] 
                    
    lucro_subprodutos = sum( (by_products_capacity.loc[v]['price'] - by_products_capacity.loc[v]['cost'])*model.q_sub[v,t] 
                            for v in DESTINOS for t in PERIODOS)
    
    lucro_total = lucro_venda + lucro_subprodutos - custo_ativacao - custo_estoque
    
    model.restricoes = pyo.ConstraintList()
    custo_atrito = sum( (1 - model.y[pedido[0],pedido[1],pedido[2]]) * 10000 for pedido in demand.index)#considerando um custo fixo 10000 por unidade nao atendida 
    model.restricoes.add(custo_atrito == model.custo_atrito)
    
    if isMultiObjective == False:
        model.OBJ = pyo.Objective(expr = lucro_total, sense=pyo.maximize)
    else:
        model.OBJ = pyo.Objective(expr = alpha*lucro_total - beta*custo_atrito , sense=pyo.maximize)
    #Montar restrições
    model.restricoes.add(model.custo_ativacao == custo_ativacao)
    model.restricoes.add(model.custo_estoque == custo_estoque)
    model.restricoes.add(model.lucro_venda == lucro_venda)
    model.restricoes.add(model.lucro_subprodutos == lucro_subprodutos)
    
    

    for i in PRODUTOS:
        for t in PERIODOS:
            if sum(existe[j-1][i-1][t-1] for j in PEDIDOS) == 0:
                continue
            
            soma_y = sum(model.y[j,i,t] * demand.loc[j,i,t]['demand'] for j in PEDIDOS if existe[j-1][i-1][t-1] == 1)
            if t > 1:
                restricao_demanda = sum(model.I[i,u,t-1] for u in UNIDADESDEARMAZEM)+ sum( model.q[i,k,t] for k in production_arcs.loc[i].index) - sum(model.I[i,u,t] for u in UNIDADESDEARMAZEM) == soma_y
            else:
                restricao_demanda = sum( model.q[i,k,t] for k in production_arcs.loc[i].index) - sum(model.I[i,u,t] for u in UNIDADESDEARMAZEM) == soma_y 

            model.restricoes.add(restricao_demanda)
    
    for k in MAQUINAS:
        for t in PERIODOS:
            restricao_ativacao = sum( model.q[i,k,t] for i in production_arcs.xs(k, level = 1).index ) <= BIG_M * model.w[k,t]
            model.restricoes.add(restricao_ativacao)  
            
            restricao_capacidade_producao = sum( production_arcs.loc[i,k]['eficiency'] * model.q[i,k,t] for i in production_arcs.xs(k, level = 1).index )
            if hasLimitedProduction == True:
                model.restricoes.add(restricao_capacidade_producao <= 60 * production_capacity.loc[k]['capacity'])
            else:
                model.restricoes.add(restricao_capacidade_producao <= BIG_M * production_capacity.loc[k]['capacity'])
    
    
    for t in PERIODOS:
        for l in UNIDADESDEARMAZEM:
            model.restricoes.add(model.I_L[l,t] == sum( model.I[i,l,t] for i in PRODUTOS))
            
            if hasLimitedStock == True:
                model.restricoes.add(model.I_L[l,t] <= model.z[l,t] * storage_capacity.loc[l]['capacity'])
            else:
                model.restricoes.add(model.I_L[l,t] <= model.z[l,t] * BIG_M)
                continue
        
            
            
        restricao_q_sub =sum(model.q_sub[v,t] for v in DESTINOS)== sum(by_products.loc[i,k]['by_product'] * model.q[i,k,t] for i in PRODUTOS for k in production_arcs.loc[i].index)
            
        model.restricoes.add(restricao_q_sub)
        for v in DESTINOS:
            model.restricoes.add(model.q_sub[v,t] <= by_products_capacity.loc[v]['capacity'])
    
    return model
```

### Resolvendo o modelo  inicial padrão


```python
model = MontarModelo(demand,production_arcs,production_capacity,
             cost_revenues,by_products_capacity,by_products,activation_cost,storage_capacity)

#setando o solver cplex
solver = pyo.SolverFactory('cplex')
#limitador o tempo para 1min
solver.options['timelimit'] = 60

#resolver o modelo em si
results = solver.solve(model) #, tee=True)
print(results)
#model.pprint()
```

No período/mês 1, as linhas de produção 1,2,3,5,6 foram usadas, assim por diante.

## Analizando custos de atrito com clientes

```python
### Montando frente de Pareto variando alfa e beta
import numpy as np

Pareto_Optimals = []

for alpha in np.arange(0.05,1,0.05):
    beta = 1 - alpha
    model = MontarModelo(demand,production_arcs,production_capacity,
                 cost_revenues,by_products_capacity,by_products,activation_cost,storage_capacity, True, True, True, alpha, beta)

    #setando o solver cplex
    solver = pyo.SolverFactory('cplex')
    #limitador o tempo para 1min
    solver.options['timelimit'] = 60

    #resolver o modelo em si
    results = solver.solve(model) #, tee=True)
    Pareto_Optimals.append([alpha, beta,
                            model.lucro_venda.value - model.custo_ativacao.value - model.custo_estoque.value + model.lucro_subprodutos.value,
                            model.custo_atrito.value])
```

```python
Pareto_Optimals = pd.DataFrame(Pareto_Optimals, columns = ['alpha', 'beta', 'Lucro', 'Atrito'])
Pareto_Optimals = Pareto_Optimals.set_index(['alpha','beta'])
print(Pareto_Optimals)

import matplotlib.pyplot as plt
```

```python
plt.scatter(Pareto_Optimals['Lucro'], Pareto_Optimals['Atrito'], marker=r'$\clubsuit$',
            label="Luck")
```

##### A análise acabou não indicando muitos valores de custos de atrito diferentes para montar uma frente de Pareto mais robusta, mas mesmo com dois valores, é possível observar que o ganho é maior quando o custo de atrito é maior.

## Resolvendo modelo inicial novamente

```python
model = MontarModelo(demand,production_arcs,production_capacity,
             cost_revenues,by_products_capacity,by_products,activation_cost,storage_capacity)

#setando o solver cplex
solver = pyo.SolverFactory('cplex')
#limitador o tempo para 1min
solver.options['timelimit'] = 60

#resolver o modelo em si
results = solver.solve(model) #, tee=True)
print(results)
#results.display()
#model.pprint()
```

O gap final da solução foi de $0.00%$ o que indica que a solução é ótima. Ele pode ser calculado como a diferença percentual da melhor solução do problema com o maior limite inferior encontrado de uma relaxação. Ou seja, (upper bound - lower bound)/lowerbound.

Ele indica o quão próximo de uma prova de otimalidade a melhor solução está.

O número de nós de branch em bound final foi de $11$. O que indica que desde o nó raíz, a árvore de branch em bound realizou $11$ "branches" de subproblemas relaxados até encontrar uma solução inteira viável com prova de otimalidade.

O tempo total de execução foi de $0.09$ segundos. O que foi bem rápido, é possível que a instância do problema seja bem pequena ou que haja algum furo no modelo que não consegui notar até o momento desse relatório o que me limitou avaliação do comportamento do tempo de execução do modelo com uso de menos variáveis e restrições.

O valor final de função objetivo foi de $299583.70548$.

Caso o modelo não seja escalonável para instâncias maiores e o tempo de execução disponível seja impraticável para obter soluções viáveis e boas, uma opção para resolvê-lo é o uso de heurísticas.

Para problemas de dimensionamento de lotes como esse, é bastante usado heurísticas do tipo relax-and-fix. Que consiste em resolver subproblemas relaxados e fixando valores inteiros para os subproblemas seguintes. Várias estratégias de criar esses subproblemas podem ser avaliadas.
